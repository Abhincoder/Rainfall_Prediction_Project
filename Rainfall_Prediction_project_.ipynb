```python
for col in data.columns:

  # Checking if the column contains
  # any null values
  if data[col].isnull().sum() > 0:
    val = data[col].mean()
    data[col] = data[col].fillna(val)

data.isnull().sum().sum()
```

```python
data.replace({'yes':1, 'no':0}, inplace=True)
```

```python
## checking correlation
plt.figure(figsize=(10,10))
sb.heatmap(data.corr() > 0.8,
           annot=True,
           cbar=False)
plt.show()
```

```python
##we will remove the highly correlated features ‘maxtemp’ and ‘mintemp’
##temp and dewpoint provide distinct information regarding the weather and atmospheric conditions.
```

```python
data.drop(['maxtemp', 'mintemp'], axis=1, inplace=True)
```

```python
features = data.drop(['day', 'rainfall'], axis=1)
target = data.rainfall
```

```python
X_train, X_val, \
    Y_train, Y_val = train_test_split(features,
                                      target,
                                      test_size=0.2,
                                      stratify=target,
                                      random_state=2)

# As the data was highly imbalanced we will
# balance it by adding repetitive rows of minority class.
ros = RandomOverSampler(sampling_strategy='minority',
                        random_state=22)
X, Y = ros.fit_resample(X_train, Y_train)
```

```python
# Normalizing the features for stable and fast training.
scaler = StandardScaler()
X = scaler.fit_transform(X)
X_val = scaler.transform(X_val)
```

```python
# Now let’s train some state-of-the-art models for classification and train them on our training data.

# LogisticRegression
# XGBClassifier
# SVC
```

```python
models = [LogisticRegression(), XGBClassifier(), SVC(kernel='rbf', probability=True)]

for i in range(3):
  models[i].fit(X, Y)

  print(f'{models[i]} : ')

  train_preds = models[i].predict_proba(X)
  print('Training Accuracy : ', metrics.roc_auc_score(Y, train_preds[:,1]))

  val_preds = models[i].predict_proba(X_val)
  print('Validation Accuracy : ', metrics.roc_auc_score(Y_val, val_preds[:,1]))
  print()
```

```python
#  plot the confusion matrix as well for the validation data using the SVC model.
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn import metrics

ConfusionMatrixDisplay.from_estimator(models[2], X_val, Y_val)
plt.show()
```

```python
# True Negatives (TN): 16 — correctly predicted 0's.
# False Positives (FP): 8 — predicted 1 but actual was 0.
# False Negatives (FN): 3 — predicted 0 but actual was 1.
# True Positives (TP): 47 — correctly predicted 1's.
```

```python
print(metrics.classification_report(Y_val,
                                    models[2].predict(X_val)))
```

```python
# Analysis:
# Precision for Class 0: 0.84
# Precision is good, meaning when the model predicts 0, it's correct 84% of the time.

# Recall for Class 0: 0.67
# Recall is lower for class 0, meaning that out of all actual 0's, only 67% are being correctly classified as 0. This suggests the model is misclassifying more true negatives as false positives.

# F1-Score for Class 0: 0.74
# The F1-score balances precision and recall, which is useful when there's an imbalance between precision and recall.

# Class 1 Performance:
# For class 1 (rainfall predicted), the model performs much better with high precision, recall, and F1-score (0.85, 0.94, 0.90). This indicates that your model is better at predicting class 1.

# Overall Accuracy: 85%
# This indicates that 85% of the predictions made by the SVC on the validation set are correct.
```
